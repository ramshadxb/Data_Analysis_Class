```python
tips = sns.load_dataset('tips')
sns.boxplot(x='day',y='total_bill', data=tips, color='red')
plt.title("Total_Bill by Day")
plt.show()
```

```python
sns.scatterplot(x="tip",y='total_bill',data=tips,hue="sex",style="smoker")
plt.title("Total Bill vs Tip")
plt.show()
```
