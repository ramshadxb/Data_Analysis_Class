```python
#Line Plot of monthly profits

months = ["Jan", "Feb", "Mar", "Apr", "May"]
profits = [2000, 2500, 2100, 3000, 2800]

plt.plot(months,profits,marker='o',color='blue')
plt.title("Monthly Profits")
plt.xlabel("Months")
plt.ylabel("Profits")
plt.grid(True)
plt.show()
```

```python
students = ["Asha", "Ravi", "Sneha", "John", "Kavya"]
marks = [85, 78, 92, 70, 88]

plt.bar(students, marks, color=['red','blue','green','orange','purple'])
plt.title("Student Marks Comparison")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.show()
```

```python
sales = [500, 600, 750, 400, 650]
marks = [85, 78, 92, 70, 88]


plt.scatter(sales, marks, color='red')
plt.title("Sales vs Marks")
plt.xlabel("Sales")
plt.ylabel("Marks")
plt.grid(True)
plt.show()
```

```python
plt.pie(marks,labels=students,autopct='%1.1f%%',startangle=90)
plt.title("Percentage of Marks by Students")
plt.show()
```
