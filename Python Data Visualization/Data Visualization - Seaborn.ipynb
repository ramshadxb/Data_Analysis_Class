```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset("fmri")

sns.lineplot(data=df, x="timepoint", y="signal", hue="event")
plt.title("FMRI Signal Over Time")
plt.show()
```

```python
tips = sns.load_dataset("tips")

sns.barplot(x="day", y="total_bill", data=tips)
plt.title("Average Bill Amount per Day")
plt.show()
```

```python
sns.countplot(x="day", data=tips)
plt.title("Number of Records per Day")
plt.show()
```

```python
sns.boxplot(x="day", y="total_bill", data=tips)
plt.title("Box Plot of Total Bill by Day")
plt.show()
```

```python
sns.scatterplot(x="total_bill", y="tip", data=tips, hue="sex", style="smoker")
plt.title("Tips vs Total Bill")
plt.show()
```

```python
# Load the iris dataset
iris = sns.load_dataset("iris")

# Select only numeric columns for correlation
# This excludes the 'species' column which contains strings
numeric_iris = iris.select_dtypes(include=['float64', 'int64'])

# Calculate correlation matrix
corr = numeric_iris.corr()

# Create heatmap
sns.heatmap(corr, annot=True, cmap="YlGnBu")
plt.title("Correlation Matrix")
plt.show()
```
