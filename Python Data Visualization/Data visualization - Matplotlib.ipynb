```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4, 5]
y = [10, 12, 8, 14, 11]
```

```python
plt.plot(x,y,marker = 'o',color = 'green')
plt.title('Sales over trend')
plt.xlabel('Days')
plt.ylabel('Sales')
plt.grid(True)
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y1 = [10, 12, 8, 14, 11]
y2 = [9, 11, 10, 13, 12]

plt.plot(x, y1, label='Product A', marker='o')
plt.plot(x, y2, label='Product B', marker='s')
plt.title("Sales Comparison")
plt.xlabel("Day")
plt.ylabel("Sales")
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.plot(x, y, color='blue', marker='s', linestyle='--', label='Product A')
plt.title("Product A Sales Over Days")
plt.xlabel("Days")
plt.ylabel("Sales")
plt.legend()
plt.grid(True)
plt.show()
```
